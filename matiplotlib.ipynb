```python
import numpy as np
import matplotlib.pyplot as  plt
```

```python
x= np.linspace(-6,5,1500)
y=x**3 + 3*x + 7
y1= 3*x**2 + 3
y2= 6*x
print(x)
fig , ax = plt.subplots()
ax.plot(x,y,color = "red",label ="y(x)")
ax.plot(x,y1,color = "black",label ="y(x)'")
ax.plot(x,y2,color = "pink",label ="y(x)''")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.legend()
```
